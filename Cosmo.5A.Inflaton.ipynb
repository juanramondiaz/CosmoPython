### 1.1.1. M√©trica FLRW Plana
---
La m√©trica FLRW para un universo espacialmente plano (curvatura $ k = 0 $) es:
$$ds^2 = -dt^2 + a(t)^2 (dx^2 + dy^2 + dz^2),$$
donde $ a(t) $ es el factor de escala, y el par√°metro de Hubble es $ H = \dot{a}/a $, con el punto denotando derivada respecto a $ t $.

### 1.1.2. Acci√≥n y Tensor de Energ√≠a-Momento
---
El modelo considera un campo escalar $ \phi(t) $ homog√©neo (inflaton) con potencial $ V(\phi) $. La acci√≥n es:
$$S = \int d^4x \sqrt{-g} \left[ \frac{M_{Pl}^2}{2} R - \frac{1}{2} \partial_\mu \phi \partial^\mu \phi - V(\phi) \right],$$
donde $ R $ es la curvatura escalar.
Variando respecto a la m√©trica, obtenemos las ecuaciones de Einstein:
$$G_{\mu\nu} = \frac{1}{M_{Pl}^2} T_{\mu\nu},$$
con el tensor de energ√≠a-momento del campo escalar:
$$T_{\mu\nu} = \partial_\mu \phi \partial_\nu \phi - g_{\mu\nu} \left( \frac{1}{2} \partial^\lambda \phi \partial_\lambda \phi + V(\phi) \right).$$
Para $ \phi = \phi(t) $ (homog√©neo), los componentes relevantes son:

 - Densidad de energ√≠a: $ \rho = T_{00} = \frac{1}{2} \dot{\phi}^2 + V(\phi) $,
 - Presi√≥n: $ p = \frac{1}{3} T_{ii}/a^2 = \frac{1}{2} \dot{\phi}^2 - V(\phi) $ (para $ i=1,2,3 $).

**Nota**: El campo se comporta como un fluido perfecto con ecuaci√≥n de estado $ w = p/\rho $, que var√≠a de $ w \approx -1 $ (dominada por potencial) a $ w \approx 1 $ (dominada por cin√©tica).

### 1.1.3. Ecuaciones de Friedmann
---
Las ecuaciones de Einstein en m√©trica FLRW dan las ecuaciones de Friedmann. La primera (componente 00):
$$H^2 = \left( \frac{\dot{a}}{a} \right)^2 = \frac{1}{3 M_{Pl}^2} \rho = \frac{1}{3 M_{Pl}^2} \left( \frac{1}{2} \dot{\phi}^2 + V(\phi) \right).$$
La segunda ecuaci√≥n (de aceleraci√≥n, combinando componentes):
$$\frac{\ddot{a}}{a} = \dot{H} + H^2 = -\frac{1}{6 M_{Pl}^2} (\rho + 3p).$$
Sustituyendo $ \rho + 3p = 2\dot{\phi}^2 - 2V(\phi) $:
$$\dot{H} + H^2 = -\frac{1}{6 M_{Pl}^2} (2\dot{\phi}^2 - 2V) = -\frac{1}{3 M_{Pl}^2} (\dot{\phi}^2 - V).$$
Para obtener $ \dot{H} $ directamente, usamos la ecuaci√≥n de continuidad (de conservaci√≥n de $ T^{\mu\nu} $):
$$\dot{\rho} + 3H (\rho + p) = 0.$$
Sustituyendo $ \rho + p = \dot{\phi}^2 $:
$$\dot{\rho} = -3H \dot{\phi}^2.$$
Ahora, diferencia la primera ecuaci√≥n de Friedmann:
$$2 H \dot{H} = \frac{1}{3 M_{Pl}^2} \dot{\rho} = \frac{1}{3 M_{Pl}^2} (-3H \dot{\phi}^2) = -\frac{H \dot{\phi}^2}{M_{Pl}^2}.$$
Dividiendo por $ 2H $ (asumiendo $ H \neq 0 $):
$$\dot{H} = -\frac{\dot{\phi}^2}{2 M_{Pl}^2}.$$

### 1.1.4. Ecuaci√≥n de Klein-Gordon para el Inflaton
---
Variando la acci√≥n respecto a $ \phi $, obtenemos la ecuaci√≥n de movimiento:
$$\Box \phi + V'(\phi) = 0,$$
donde $ \Box = \nabla_\mu \nabla^\mu $ y $ V' = dV/d\phi $.
En m√©trica FLRW, para $ \phi(t) $ homog√©neo:
$$\ddot{\phi} + 3 H \dot{\phi} + V'(\phi) = 0.$$
(El t√©rmino $ 3H \dot{\phi} $ surge del factor de expansi√≥n, como fricci√≥n de Hubble.)

$$H^2 = \frac{1}{3 M_{Pl}^2} \left( \frac{1}{2} \dot{\phi}^2 + V(\phi) \right), \quad \dot{H} = -\frac{\dot{\phi}^2}{2 M_{Pl}^2}, \quad \ddot{\phi} + 3 H \dot{\phi} + V'(\phi) = 0.$$

### 1.1.5. Aproximaci√≥n Slow-Roll
---
La inflaci√≥n requiere expansi√≥n acelerada ($ \ddot{a} > 0 $), i.e., $ \dot{H} + H^2 > 0 $ o $ \epsilon < 1 $, donde definimos los par√°metros slow-roll:
$$\epsilon = -\frac{\dot{H}}{H^2}, \quad \eta = -\frac{\ddot{\phi}}{H \dot{\phi}}.$$
(Existen definiciones equivalentes en t√©rminos de potencial.)
En slow-roll, asumimos que el campo "rueda lentamente": $ \dot{\phi}^2 \ll V(\phi) $ (dominancia potencial) y $ |\ddot{\phi}| \ll |3 H \dot{\phi}|, |V'| $ (segunda derivada peque√±a).
De la segunda ecuaci√≥n de Friedmann:
$$\epsilon = -\frac{\dot{H}}{H^2} = \frac{\dot{\phi}^2}{2 M_{Pl}^2 H^2} = \frac{3 \dot{\phi}^2}{2 (\dot{\phi}^2 + 2V)} \approx \frac{3 \dot{\phi}^2}{4V} \ll 1.$$
Aproximando las ecuaciones:

De Klein-Gordon: $ 3 H \dot{\phi} + V' \approx 0 $ ‚áí $ \dot{\phi} \approx - \frac{V'}{3H} $.
De primera Friedmann: $ H^2 \approx \frac{V}{3 M_{Pl}^2} $.

Sustituyendo:
$$\dot{\phi}^2 \approx \frac{(V')^2}{9 H^2} \approx \frac{(V')^2}{3} \cdot \frac{3 M_{Pl}^2}{V} = M_{Pl}^2 \frac{(V')^2}{V}.$$
Entonces:
$$\epsilon \approx \frac{M_{Pl}^2}{2} \left( \frac{V'}{V} \right)^2, \quad \eta \approx M_{Pl}^2 \left( \frac{V''}{V} - \frac{1}{2} \left( \frac{V'}{V} \right)^2 \right).$$
El n√∫mero de e-folds $ N = \int H dt \approx -\frac{1}{M_{Pl}^2} \int_{\phi_i}^{\phi_f} \frac{V}{V'} d\phi $, y la inflaci√≥n termina cuando $ \epsilon \approx 1 $.
Estas aproximaciones permiten calcular observables como el √≠ndice espectral $ n_s \approx 1 - 6\epsilon + 2\eta $ y la raz√≥n tensor-escalar $ r \approx 16\epsilon $, comparables con datos del CMB.

| Magnitud     | F√≥rmula (slow-roll)                                 | Interpretaci√≥n                      |
| ------------ | --------------------------------------------------- | ----------------------------------- |
| $\epsilon_V$ | $\frac{1}{2}\bar M_{\rm Pl}^2 (V'/V)^2$             | mide inclinaci√≥n del potencial      |
| $\eta_V$     | $\bar M_{\rm Pl}^2 V''/V$                           | mide curvatura del potencial        |
| $n_s$        | $1 - 6\epsilon_V + 2\eta_V$                         | √≠ndice espectral escalar            |
| $r$          | $16\epsilon_V$                                      | relaci√≥n tensor/escalar             |
| Evaluar en   | $\phi_*$ cuando $N_{\text{end}} - N_* \simeq 50-60$ | salida del horizonte del modo pivot |

### 1.2.1. Potencial Cuadr√°tico (Quad) - Chaotic Inflation
---
**Potencial Cudr√°tico**$$V(\phi) = \frac{1}{2} m^2 \phi^2$$
Derivada:
$$V'(\phi) = m^2 \phi$$

Descripci√≥n F√≠sica

- **Modelo**: Inflaci√≥n ca√≥tica, propuesto por Linde (1983).
- **Caracter√≠sticas**:
- - Potencial simple de tipo "large-field" (campo grande), donde $ \phi \gg M_{Pl} $ inicialmente.
- - Genera inflaci√≥n cuando el t√©rmino potencial domina ($ \frac{1}{2} \dot{\phi}^2 \ll V(\phi) $).
- - Par√°metro $ m $ (masa del inflaton) se ajusta a $ m \approx 6 \times 10^{-6} M_{Pl} $ para coincidir con la amplitud de fluctuaciones del CMB ($ A_s \approx 2.1 \times 10^{-9} $).

**Par√°metros Slow-Roll:**

- $ \epsilon_v \approx \frac{M_{Pl}^2}{2} \left( \frac{V'}{V} \right)^2 = \frac{2 M_{Pl}^2}{\phi^2} $
- $ \eta_v \approx M_{Pl}^2 \frac{V''}{V} = \frac{2 M_{Pl}^2}{\phi^2} $
- N√∫mero de e-folds: $ N \approx \frac{\phi_0^2 - \phi_f^2}{4 M_{Pl}^2} $, con fin de inflaci√≥n en $ \phi_f \approx \sqrt{2} M_{Pl} $.
- Observables: $ n_s \approx 1 - \frac{4 M_{Pl}^2}{\phi_0^2} $, $ r \approx \frac{32 M_{Pl}^2}{\phi_0^2} $.
- Valores Esperados para $ \phi_0 \approx 15 M_{Pl} $, $ n_s \approx 0.964 $, $ r \approx 0.142 $.

**Comentarios**
- Ventajas: Simple de implementar y analizar. Bien estudiado.
- Limitaciones: El valor de $ r $ es alto comparado con l√≠mites de Planck 2018 ($ r < 0.036 $), lo que lo hace marginalmente desfavorecido.
- Uso en el C√≥digo: $ m = \exp(\log m) $ asegura que $ m $ sea positivo, √∫til para optimizaci√≥n num√©rica.


### 1.2.2. Potencial Cu√°rtico (Quartic)
---
**Potencial Cu√°rtico**
$$V(\phi) = \frac{1}{4} \lambda \phi^4$$
**Derivada:**
$$V'(\phi) = \lambda \phi^3$$
**Descripci√≥n F√≠sica**

- **Modelo**: Otro ejemplo de inflaci√≥n ca√≥tica de "large-field".
- **Caracter√≠sticas:**
- - Potencial de potencia superior ($ \phi^4 $), m√°s empinado que el cuadr√°tico.
- - $ \lambda $ es un acoplamiento adimensional, t√≠picamente $ \lambda \approx 10^{-13} $ para ajustar $ A_s $  a los datos del CMB.
- - Genera mayor contribuci√≥n de modos tensoriales (ondas gravitacionales).

**Par√°metros Slow-Roll:**
- $ \epsilon_v \approx \frac{M_{Pl}^2}{2} \left( \frac{V'}{V} \right)^2 = \frac{8 M_{Pl}^2}{\phi^2} $
- $ \eta_v \approx M_{Pl}^2 \frac{V''}{V} = \frac{6 M_{Pl}^2}{\phi^2} $
- Fin de inflaci√≥n: $ \phi_f \approx 2 \sqrt{2} M_{Pl} $.
- $ N \approx \frac{\phi_0^2 - \phi_f^2}{8 M_{Pl}^2} $.
- Observables: $ n_s \approx 1 - \frac{6 M_{Pl}^2}{\phi_0^2} $, $ r \approx \frac{64 M_{Pl}^2}{\phi_0^2} $.
- Para $ \phi_0 \approx 15 M_{Pl} $, $ n_s \approx 0.973 $, $ r \approx 0.284 $.

**Comentarios**
- Ventajas: Similar al cuadr√°tico, pero produce un $ r $ mayor, √∫til para estudiar tensiones con datos de ondas gravitacionales.
- Limitaciones: Est√° m√°s desfavorecido por Planck debido a su alto $ r $.


### 1.2.3. Potencial Hilltop
---
**Potencial Hilltop**
$$V(\phi) = V_0 \left( 1 - \left( \frac{\phi}{\mu} \right)^p \right)^2$$
**Derivada:**
$$V'(\phi) = -2 V_0 \left( 1 - \left( \frac{\phi}{\mu} \right)^p \right) \frac{p \phi^{p-1}}{\mu^p}$$
- M√°ximo en $\phi \approx 0$ (para $p$ par).Inflaci√≥n ocurre cerca de la cima.
- Par√°metros t√≠picos: $ùëù=2, \,4;\,\mu\lesssim \mathcal{O}(M_{Pl})$

**Descripci√≥n F√≠sica**
- **Modelo**: Inflaci√≥n de tipo "small-field", donde $ \phi \ll M_{Pl} $, y el campo rueda desde cerca del m√°ximo del potencial ($ \phi \approx 0 $).
- **Caracter√≠sticas:**
- - $ V_0 $: Escala de energ√≠a del potencial.
- - $ \mu $: Escala del campo que determina el ancho del potencial.
- - $ p $: Exponente que controla la forma (t√≠picamente $ p = 2, 4 $).
- - La inflaci√≥n ocurre cerca de $ \phi = 0 $, donde el potencial es casi plano.

**Par√°metros Slow-Roll:**

Dependen de $ \phi/\mu $ y $ p $. Para $ p = 2 $:

- $ \epsilon_v \approx \frac{M_{Pl}^2}{2} \left( \frac{2 p (\phi/\mu)^{p-1}}{\mu (1 - (\phi/\mu)^p)} \right)^2 $
- $ \eta_v $: M√°s complejo, pero dominado por $ V'' $.
- $ N \approx \frac{V_0}{M_{Pl}^2} \int_{\phi_f}^{\phi_0} \frac{d\phi}{V'(\phi)} $, evaluado num√©ricamente.
- Puede producir $ r $ peque√±o, m√°s compatible con Planck.

**Comentarios**

- Ventajas: Modelos hilltop son favorecidos por datos recientes debido a su bajo $ r $.
- √ötil para estudiar inflaci√≥n en regiones de campo peque√±o.
- Limitaciones: C√°lculos anal√≠ticos son m√°s complejos; requiere simulaci√≥n num√©rica precisa.


### 1.2.4. Potencial $\alpha$-atractor (Starobinsky-like)
---
**Potencial Alpha (Starobinsky-like)**
$$V(\phi) = V_0 \left( 1 - e^{-\sqrt{\frac{2}{3\alpha}} \phi} \right)^2$$
**Derivada:**
$$V'(\phi) = 2 V_0 \left( 1 - e^{-\sqrt{\frac{2}{3\alpha}} \phi} \right) \sqrt{\frac{2}{3\alpha}} e^{-\sqrt{\frac{2}{3\alpha}} \phi}$$
**Descripci√≥n F√≠sica**

- **Modelo**: Inspirado en inflaci√≥n tipo Starobinsky o modelos de $\alpha$-atractores, comunes en teor√≠a de cuerdas.
- **Caracter√≠sticas:**

- - $ V_0 $: Escala de energ√≠a.
- - $ \alpha $: Par√°metro que controla la curvatura del potencial (t√≠picamente $ \alpha \sim 1 $).
- - El t√©rmino exponencial hace que el potencial sea plano para $ \phi \gg \sqrt{\frac{3\alpha}{2}} M_{Pl} $, favoreciendo inflaci√≥n.

**Par√°metros Slow-Roll:**
- $ \epsilon_v \approx \frac{4 M_{Pl}^2}{3\alpha (e^{\sqrt{\frac{2}{3\alpha}} \phi} - 1)^2} $
- $ \eta_v $: Negativo y peque√±o para $ \phi $ grande.
- $ N \approx \frac{3\alpha}{4 M_{Pl}^2} (e^{\sqrt{\frac{2}{3\alpha}} \phi_0} - e^{\sqrt{\frac{2}{3\alpha}} \phi_f}) $.
- Produce $ r $ muy peque√±o, compatible con observaciones.

**Comentarios**

- Ventajas: Muy favorecido por datos (bajo $ r $), relevante para modelos modernos.
- Limitaciones: Requiere cuidado num√©rico debido al t√©rmino exponencial.

### 2.1.1. Par√°metros F√≠sicos y de la Simulaci√≥n
---
- Masa de Planck:        $\quad M_{Pl} = 2.44 \cdot 10^{18}$  (GeV)
- Masa del Inflaton:     $\quad m = 6\cdot10^{-6}\,M_Pl$
- Los valores de $ m $, $ \lambda $, y $ V_0 $ est√°n ajustados para $ A_s \approx 2.1 \times 10^{-9} $. Pueden recalcularse usando:
$$A_s \approx \frac{V}{24 \pi^2 M_{Pl}^4 \epsilon}.$$
- Tiempo de Integraci√≥n: $\quad [0, 100/m]$
- Opciones del Potencial:  **'quad', 'quartic', 'hilltop', 'alpha'**

```python
UNITS     = 'SI' # 'SI' o 'NATURAL'
POTENTIAL = 'quad'# 'quad', 'quartic', 'hilltop', 'alpha'
```

```python
def get_Params(units=UNITS, potential=POTENTIAL, CMB_fit = True):
    """
    Calcula las constantes f√≠sicas para inflaci√≥n c√≥smica seg√∫n el sistema de unidades.
    
    Args:
        units (str): Sistema de unidades ('NATURAL' o 'SI').
        c (float, optional): Velocidad de la luz (m/s). Usado solo en SI.
        hbar (float, optional): Constante de Planck reducida (J¬∑s). Usado solo en SI.
    
    Returns:
        dict: Diccionario con constantes f√≠sicas:
            - M_Pl: Masa de Planck reducida
            - m: Masa del inflaton (para potencial cuadr√°tico)
            - lambda_: Acoplamiento (para potencial cu√°rtico)
            - V0: Escala de energ√≠a (para hilltop y alpha)
            - alpha: Par√°metro para potencial alpha
            - mu: Par√°metro para potencial hilltop
            - p: Par√°metro para potencial hilltop
            - G: Constante gravitacional
    
    Raises:
        ValueError: Si las unidades no son v√°lidas.
    """
    
    # Constantes f√≠sicas fundamentales (en SI)
    c = 2.99792458e8  # m/s
    hbar = 1.0545718e-34  # J¬∑s
    G = 6.67430e-11  # m¬≥ kg‚Åª¬π s‚Åª¬≤
    GeV = 1.60217662e-10 # J (Factor de Conversion de Unidades GeV --> J)
   
    if units == 'SI':
        argument = hbar * c / (8 * torch.pi * G)
        M_Pl = torch.sqrt(torch.as_tensor(argument))  # kg
            
    if units == 'NATURAL':        # En unidades naturales: ƒß = c = 1
        # Factores de Conversi√≥n. Se utilizan valores en GeV para consistencia con literatura de cosmolog√≠a
        hbar_GeV  = hbar/GeV       # m^3 Kg^-1 s^-2 --> GeV Seg
        metro_GeV = GeV/(hbar*c)   # metro --> GeV^-1
        Kg_GeV    = c**2/GeV       # Kg --> GeV
        seg_GeV   = GeV/hbar    # Segundos -->  GeV^-1
        # Constante G en unidades GeV^-2
        G = G*metro_GeV**3/(Kg_GeV*seg_GeV**2)  # GeV^-2
        # Redefinimos  ƒß = c = 1
        c = 1                       # Unidades Naturales
        hbar = 1                    # Unidades Naturales
        argument = hbar * c / (8 * torch.pi * G)
        M_Pl = torch.sqrt(torch.as_tensor(argument))  # GeV
        
    elif units == 'PLANCK':
        # En unidades Planck: M_Pl = 1
        hbar = 1
        c = 1        
        G = 1
        M_Pl = 1
    
    # Eneg√≠a de Planck
    E_Pl = M_Pl * c**2
    
    # Masa del inflaton 
    m_natural = 1.0
    if CMB_fit:  # m ‚âà 6e-6 M_Pl ajuste para comparar con datos CMB, A_s ‚âà 2.1e-9)
        m_natural = m_natural * 6e-6
    m = m_natural * M_Pl  # kg
    
    
    # Acoplamiento Œª (adimensional, ajustado para potencial cu√°rtico)
    lambda_ = 1e-13  # Adimensional
        
    # Escala de energ√≠a V‚ÇÄ (ajustado para modelos small-field)
    # V‚ÇÄ ‚âà 1e-10 M_Pl‚Å¥ en unidades naturales
    V0_natural = 1e-10  # Relaci√≥n V‚ÇÄ/M_Pl‚Å¥
    # En SI: V‚ÇÄ = V0_natural √ó (M_Pl c¬≤)‚Å¥ / (ƒß c)¬≥ para J/m¬≥
    V0 = V0_natural * (M_Pl*(c**2))**4  /((hbar*c)**3)   # J/m¬≥ (en SI)
    
    # Parametro mu utilizado en potencial Hilltop
    mu = 5.0*M_Pl
    # Par√°metro Œ± (adimensional)
    alpha = 1.0  # Adimensional
    
    param = {
        'hbar': torch.as_tensor(hbar),      # J¬∑s
        'c': torch.as_tensor(c),            # m/s
        'G': torch.as_tensor(G),            # m¬≥ kg‚Åª¬π s‚Åª¬≤
        'M_Pl': torch.as_tensor(M_Pl),      # kg
        'E_Pl': torch.as_tensor(E_Pl),      # J
        'm': torch.as_tensor(m),            # 
        'lambda_': torch.as_tensor(lambda_),# adimensional
        'V0': torch.as_tensor(V0),          # 
        'mu': torch.as_tensor(mu),          # 
        'p': torch.as_tensor(5.0),          # adimensional 
        'alpha': torch.as_tensor(alpha),    # adimensional
        'units': units,    # Sistema de unidades
        'potential': potential,  # Tipo de Potencial
        'use_epsilon_v':False    
    }        
   
    return param

def print_param(param):
    """
    Imprime las constantes f√≠sicas de forma legible.
    
    Args:
        constants (dict): Diccionario de constantes f√≠sicas.
        units (str): Sistema de unidades utilizado.
    """
    print(f"\n=== Constantes F√≠sicas ({param['units']}) y Par√°metros\n------------------------------------------")
    for key, value in param.items():
        if key == 'hbar':
            print(f"{key}:  {value:.3e}  {'J¬∑s' if param['units'] == 'SI' else 'hbar'}")
        elif key == 'c': 
            print(f"{key}:  {value:.3e}  {'m/s' if param['units'] == 'SI' else 'c'}")
        elif key == 'G':
            print(f"{key}:  {value:.3e}  {'m¬≥ kg‚Åª¬π s‚Åª¬≤' if param['units'] == 'SI' else 'GeV‚Åª¬≤'}")
        elif key == 'E_Pl':
            print(f"{key}:  {value:.3e}  {'J' if param['units'] == 'SI' else 'GeV'}")
        elif key in ['M_Pl', 'm', 'mu']:
            print(f"{key}:  {value:.3e}  {'kg' if param['units'] == 'SI' else 'GeV'}")
        elif key == 'V0':
            print(f"{key}:  {value:.3e}  {'J/m¬≥' if param['units'] == 'SI' else 'GeV‚Å¥'}")
        elif key in ['lambda_', 'alpha', 'p']:
            print(f"{key}:  {value} (adimensional)")
        else:
            print(f"{key}:  {value}")
        
UNITS = 'SI'
param = get_Params(units=UNITS, potential= POTENTIAL, CMB_fit = True)
print_param(param)
    
UNITS = 'NATURAL'
param = get_Params(units=UNITS, potential= POTENTIAL, CMB_fit = True)
print_param(param)

UNITS = 'PLANCK'
param = get_Params(units=UNITS, potential= POTENTIAL, CMB_fit = False)
print_param(param)
```

### 2.1.2. DeepLearning VNet Potencial
---

```python
# === Red neuronal del potencial ===
class VNet(nn.Module):
    def __init__(self, hidden=128, depth=4, Vmin=0.0, act="silu"):
        super().__init__()
        acts = {
            "silu": nn.SiLU(),
            "tanh": nn.Tanh(),
            "gelu": nn.GELU(),
        }
        activation = acts.get(act, nn.SiLU())

        layers = [nn.Linear(1, hidden), activation]
        for _ in range(depth - 1):
            layers += [nn.Linear(hidden, hidden), activation]
        layers += [nn.Linear(hidden, 1)]  # salida escalar f_Œ∏(œÜ)

        self.net = nn.Sequential(*layers)
        self.softplus = nn.Softplus(beta=1.0)  # Œ≤=1 -> forma suave y estable
        # Vmin no entrenable (si quieres fijar un offset ‚â•0)
        self.register_buffer("Vmin", torch.tensor(Vmin, dtype=torch.get_default_dtype()))

    def forward(self, phi):
        """
        phi: tensor de forma (N,1) o (1,1)
        return: V(phi) con forma (N,1)
        """
        f = self.net(phi)
        return self.softplus(f) + self.Vmin

# === Carga de pesos entrenados ===
checkpoint = torch.load("./V_inflaton_DL.pt", map_location="cpu")
VNet_model = VNet(hidden=checkpoint['config']['hidden'],
            depth=checkpoint['config']['depth'],
            Vmin=checkpoint['config']['Vmin'],
            act=checkpoint['config']['act'],
            #fourier_features=checkpoint['config']['fourier_features'],
            #n_freqs=checkpoint['config']['n_freqs'],
            #spectral_norm=checkpoint['config']['spectral_norm']
                           )
VNet_model.load_state_dict(checkpoint['model_state_dict'])
VNet_model = VNet_model.double()
VNet_model.eval()
```

```python
def V_VNet(phi):
    """
    Potencial neuronal V(œÜ) en tensor torch.
    œÜ puede ser float o tensor 1D.
    """
    if not torch.is_tensor(phi):
        phi = torch.tensor([[phi]], dtype=torch.float64)
    else:
        phi = phi.reshape(-1, 1).to(torch.float64)
    V = VNet_model(phi).item()
    return torch.tensor(V)

def dV_VNet(phi):
    """
    Primera derivada V'(œÜ) usando autograd.
    """
    phi_t = torch.tensor([[phi]], dtype=torch.float64, requires_grad=True)
    V = VNet_model(phi_t)
    Vp = autograd.grad(V, phi_t, grad_outputs=torch.ones_like(V),
                       create_graph=False, retain_graph=False)[0]
    return torch.tensor(Vp.detach().cpu()[0].item())

def d2V_VNet(phi):
    """
    Segunda derivada V''(œÜ).
    """
    phi_t = torch.tensor([[phi]], dtype=torch.float64, requires_grad=True)
    V = VNet_model(phi_t)
    Vp = autograd.grad(V, phi_t, grad_outputs=torch.ones_like(V),
                       create_graph=True, retain_graph=True)[0]
    Vpp = autograd.grad(Vp, phi_t, grad_outputs=torch.ones_like(Vp),
                        create_graph=False, retain_graph=False)[0]
    return torch.tensor(Vpp.detach().cpu()[0].item())
```

### 2.1.2. Potencial y Derivada del Potencial
---


```python
def V(phi, param):
    """
    Calcula el potencial V(phi) para un campo escalar dado el tipo de potencial.
    
    Args:
        phi (torch.Tensor): Valor del campo escalar.
        param(dict): Diccionario de Par√°metros
    
    Returns:
        torch.Tensor: Valor del potencial V(phi).
    
    Raises:
        ValueError: Si el tipo de potencial no es v√°lido.
    """
    pot = param['potential']
    phi = torch.as_tensor(phi)
    if pot == 'quad':
        m = param['m']
        return 0.5 * m**2 * phi**2
    elif pot == 'quartic':
        lam = param['lambda_']
        return 0.25 * lam * phi**4
    elif pot == 'six':
        lam = param['lambda_']
        return (1/6) * lam * phi**6
    elif pot == 'hilltop':
        V0 = param['V0']
        mu = param['mu']
        p  = param['p']
        x  = (phi / mu).pow(p)
        return V0 * (1.0 - x)**2
    elif pot == 'alpha':
        V0    = param['V0']
        alpha = max(param['alpha'], 1e-12)
        s     = torch.sqrt(torch.tensor(2.0) / (torch.tensor(3.0) * alpha))
        return V0 * (1.0 - torch.exp(-s * phi))**2
    elif pot == 'VNet':
        return V_VNet(phi)
    else:
        raise ValueError("Potencial desconocido")

def dV_dphi(phi, param):
    """
    Calcula la derivada del potencial V'(phi) = dV/dphi.
    
    Args:
        phi (torch.Tensor): Valor del campo escalar (en unidades de M_Pl).
        param(dict): Diccionario de Par√°metros
    
    Returns:
        torch.Tensor: Valor de la derivada V'(phi).
    
    Raises:
        ValueError: Si el tipo de potencial no es v√°lido.
    """
    # Tipo de potencial ('quad', 'quartic', 'hilltop', 'alpha').
    pot = param['potential']
    phi = torch.as_tensor(phi)
    if pot == 'quad':
        m = param['m']
        return (m**2) * phi
    elif pot == 'quartic':
        lam = param['lambda_']
        return lam * phi**3
    elif pot == 'six':
        lam = param['lambda_']
        return lam * phi**5
    elif pot == 'hilltop':
        V0 = param['V0']
        mu = param['mu']
        p  = param['p']
        # d/dphi [ V0 (1 - (phi/mu)^p)^2 ] = -2 V0 (1 - x) * p * phi^{p-1}/mu^p
        x  = (phi / mu).pow(p)
        return -2.0 * V0 * (1.0 - x) * p * (phi.pow(p-1)) / (mu.pow(p))
    elif pot == 'alpha':
        V0    = param['V0']
        alpha = max(param['alpha'], 1e-12)
        s     = torch.sqrt(torch.tensor(2.0) / (3.0 * alpha))
        e     = torch.exp(-s * phi)
        return 2.0 * V0 * (1.0 - e) * s * e
    elif pot == 'VNet':
        return dV_VNet(phi)
    else:
        raise ValueError("Potencial desconocido")

def d2V_dphi(phi, param):
    # Segunda derivada, √∫til para eta_V
    pot = param['potential']
    phi = torch.as_tensor(phi)
    if pot == 'quad':
        m = param['m']
        return (m**2) * torch.ones_like(torch.as_tensor(phi))
    elif pot == 'quartic':
        lam = param['lambda_']
        return 3.0 * lam * phi**2
    elif pot == 'hilltop':
        V0 = param['V0']
        mu = param['mu']
        p  = param['p']
        # d/dphi de la expresi√≥n anterior:
        # V' = -2 V0 (1 - x) p phi^{p-1} / mu^p, con x=(phi/mu)^p
        # V'' = -2 V0 [ (-x') p phi^{p-1}/mu^p + (1 - x) p (p-1) phi^{p-2}/mu^p ]
        x   = (phi / mu).pow(p)
        xp  = p * (phi.pow(p-1)) / (mu.pow(p))
        term1 = -xp * p * phi.pow(p-1) / (mu.pow(p))
        term2 = (1 - x) * p * (p-1) * phi.pow(p-2) / (mu.pow(p))
        return -2.0 * V0 * (term1 + term2)
    elif pot == 'alpha':
        V0    = param['V0']
        alpha = max(param['alpha'], 1e-12)
        s     = torch.sqrt(torch.tensor(2.0) / (3.0 * alpha))
        e     = torch.exp(-s * phi)
        # V' = 2 V0 s e (1 - e) = 2 V0 s (e - e^2)
        # V'' = 2 V0 s (-s e) - 4 V0 s (-s e) e = 2 V0 s^2 ( -e + 2 e^2 )
        return 2.0 * V0 * (s**2) * (-e + 2.0*e**2)
    elif pot == 'VNet':
        return d2V_VNet(phi)
    else:
        raise ValueError("Potencial desconocido")
```

---
Ecuaciones cosmol√≥gicas, incluyendo la ecuaci√≥n de Friedmann, los par√°metros de slow-roll ($   \epsilon_H   $, $   \epsilon_V   $, $   \eta_V   $), la ecuaci√≥n de Klein-Gordon y la condici√≥n para el fin de la inflaci√≥n

**1. ECUACI√ìN DE FRIEDMANN**
Esta funci√≥n calcula el par√°metro de Hubble $ H $, que describe la tasa de expansi√≥n del universo.

- **Densidad de energ√≠a:**
$$\rho = \frac{1}{2} \frac{\dot{\phi}^2}{c^2} + V(\phi)$$
donde:

$ \rho $: Densidad de energ√≠a
$ \dot{\phi} $: Derivada temporal del campo escalar (phi_dot)
$ c $: Velocidad de la luz
$ V(\phi) $: Potencial en $ \phi $ (calculado por $ V(\phi, param) $)

- **Factor Constantes:**
$$\text{factor}_H = \frac{\hbar}{c} \cdot \frac{1}{3 M_{\text{Pl}}^2}$$
donde:

$ \hbar $: Constante de Planck reducida (hbar)
$ M_{\text{Pl}} $: Masa de Planck


- **Ecuaci√≥n de Friedmann:**
$$H = \sqrt{\text{factor}_H \cdot \rho} = \sqrt{\frac{\hbar}{c} \cdot \frac{1}{3 M_{\text{Pl}}^2} \cdot \left( \frac{1}{2} \frac{\dot{\phi}^2}{c^2} + V(\phi) \right)}$$


**2. Par√°metro de Slow-Roll $ \epsilon_H $ (epsilon_H_Equation)**
Esta funci√≥n calcula el par√°metro de slow-roll $ \epsilon_H $, que cuantifica la tasa de cambio del par√°metro de Hubble.

- Factor de epsilon:
$$\text{factor}_\epsilon = \frac{1}{2} \cdot \frac{\hbar}{c^3} \cdot \frac{1}{M_{\text{Pl}}^2}$$

- Par√°metro de slow-roll $ \epsilon_H $:
$$\epsilon_H = \text{factor}_\epsilon \cdot \frac{\dot{\phi}^2}{H^2} = \frac{1}{2} \cdot \frac{\hbar}{c^3} \cdot \frac{1}{M_{\text{Pl}}^2} \cdot \frac{\dot{\phi}^2}{H^2}$$

**3. Par√°metro de Slow-Roll $ \epsilon_V $ (epsilon_V_Equation)**
Esta funci√≥n calcula el par√°metro de slow-roll basado en el potencial, $ \epsilon_V $.

- **Potencial y su derivada:**

- - $ V(\phi) $: Potencial en $ \phi $ (calculado por $ V(\phi, param) $)
- - $ \frac{dV}{d\phi} $: Primera derivada del potencial respecto a $ \phi $ (calculada por $ dV_dphi(\phi, param) $)


- **Factor de epsilon:**
$$\text{factor}_\epsilon = \frac{1}{2} M_{\text{Pl}}^2$$

- **Par√°metro de slow-roll $ \epsilon_V $:**
$$\epsilon_V = \text{factor}_\epsilon \cdot \left( \frac{\frac{dV}{d\phi}}{V(\phi)} \right)^2 = \frac{1}{2} M_{\text{Pl}}^2 \cdot \left( \frac{\frac{dV}{d\phi}}{V(\phi)} \right)^2$$


**4. Par√°metro de Slow-Roll $ \eta_V $ (eta_V_Equation)**
Esta funci√≥n calcula el segundo par√°metro de slow-roll $ \eta_V $, relacionado con la curvatura del potencial.

- **Potencial y su segunda derivada:**

- - $ V(\phi) $: Potencial en $ \phi $
- - $ \frac{d^2V}{d\phi^2} $: Segunda derivada del potencial respecto a $ \phi $ (calculada por $ d2V_dphi(\phi, param) $)

- **Factor de eta:**
$$\text{factor}_\eta = M_{\text{Pl}}^2$$

- **Par√°metro de slow-roll $ \eta_V $:**
$$\eta_V = \text{factor}_\eta \cdot \frac{\frac{d^2V}{d\phi^2}}{V(\phi)} = M_{\text{Pl}}^2 \cdot \frac{\frac{d^2V}{d\phi^2}}{V(\phi)}$$


**5. Ecuaciones de Movimiento (equations)**
Esta funci√≥n define el sistema de ecuaciones diferenciales para el campo escalar $ \phi $ y su derivada temporal $ \dot{\phi} $.

- **Entrada:**

- - $ y = [\phi, \dot{\phi}] $: Vector de estado que contiene el campo escalar y su primera derivada temporal
- - $ \frac{dV}{d\phi} $: Primera derivada del potencial (calculada por $ dV_dphi(\phi, param) $)
- - $ H $: Par√°metro de Hubble de la ecuaci√≥n de Friedmann

- **Ecuaci√≥n de Klein-Gordon:**
$$\ddot{\phi} = -3 H \dot{\phi} - c^2 \frac{dV}{d\phi}$$

- **Sistema de ecuaciones:**
$$\frac{d}{dt} \begin{bmatrix} \phi \\ \dot{\phi} \end{bmatrix} = \begin{bmatrix} \dot{\phi} \\ -3 H \dot{\phi} - c^2 \frac{dV}{d\phi} \end{bmatrix}$$


**6. Condici√≥n de Fin de la Inflaci√≥n (end_inflation)**
Esta funci√≥n define la condici√≥n para el fin de la inflaci√≥n, cuando el par√°metro de slow-roll $ \epsilon $ alcanza el valor 1.

- **Par√°metro de slow-roll:**
$$\epsilon = \min(\epsilon_V, \epsilon_H)$$
donde:

- $ \epsilon_V $: De epsilon_V_Equation
$-  \epsilon_H $: De epsilon_H_Equation


- **Condici√≥n de fin de la inflaci√≥n:**
$$\epsilon - 1 = 0$$

- La integraci√≥n se detiene cuando $ \epsilon = 1 $.
- La condici√≥n est√° configurada para detectar la transici√≥n de $ \epsilon < 1 $ a $ \epsilon > 1 $ (direction = 1)

```python
# Ecuaciones de movimiento (Klein-Gordon + Friedmann)

def friedmann_Equation(phi, phi_dot, param):
    hbar = param['hbar']
    c = param['c']
    M_Pl = param['M_Pl']
    
    V_phi = V(phi, param)                               # Potencial en Phi
    energy_density = 0.5 * phi_dot**2/c**2 + V_phi      # Densidad de Energ√≠a
    factor_H = (hbar/c) /(3 * M_Pl**2)                  # Factor de la ecuaci√≥n de Friedmann
    
    H = torch.sqrt(factor_H * energy_density)              # Ecuaci√≥n de Friedmann

    return H

def epsilon_H_Equation(H, phi_dot, param):
    hbar = param['hbar']
    c = param['c']
    M_Pl = param['M_Pl']
    
    factor_epsilon = 0.5 *(hbar/c**3)/ M_Pl**2         # Factor Par√°metro Epsilon_H
    epsilon_H = factor_epsilon * phi_dot**2 / H**2       # Par√°metro slow-roll

    return epsilon_H

def epsilon_V_Equation( phi, param):
    hbar = param['hbar']
    c = param['c']
    M_Pl = param['M_Pl']
    
    V_phi = V(phi, param)                               # Potencial en Phi
    dV_dphi_ = dV_dphi(phi, param)                      # Derivada del potencial respecto de Phi en Phi
    
    factor_epsilon = 0.5 * M_Pl**2                  # Factor Par√°metro Epsilon_V
        
    epsilon_V = factor_epsilon * (dV_dphi_/V_phi)**2    # Par√°metro slow-roll

    return epsilon_V

def eta_V_Equation( phi, param):
    hbar = param['hbar']
    c = param['c']
    M_Pl = param['M_Pl']
    
    V_phi = V(phi, param)                               # Potencial en Phi
    d2V_dphi_ = d2V_dphi(phi, param)                    # Derivada Segunda del potencial respecto de Phi en Phi
    
    factor_eta =  M_Pl**2                  # Factor Par√°metro Epsilon_V
        
    eta_V = factor_eta * (d2V_dphi_/V_phi)     # Parametro Eta_V

    return eta_V

def equations(t, y, param):
    phi, phi_dot = y
    c = param['c']
    dV_dphi_ = dV_dphi(phi, param)                   # Derivada del potencial respecto de Phi en Phi
    
    H = friedmann_Equation(phi, phi_dot, param)      # Ecuaci√≥n de Friedmann
    phi_ddot = -3 * H * phi_dot - (c**2)*dV_dphi_    # Ecuaci√≥n de Klein-Gordon
    return [phi_dot, phi_ddot]

# Funci√≥n para detectar el fin de la inflaci√≥n (epsilon = 1)
def end_inflation(t, y, param):
    phi, phi_dot = y
        
    H = friedmann_Equation(phi, phi_dot, param)          # Ecuaci√≥n de Friedmann
    epsilon_V = epsilon_V_Equation(phi, param)    # Par√°metro slow-roll
    epsilon_H = epsilon_H_Equation(H, phi_dot, param)      # Par√°metro slow-roll
    epsilon =torch.min(torch.Tensor(epsilon_V), torch.Tensor(epsilon_H))
    return epsilon - 1  # Cruza 0 cuando epsilon = 1

end_inflation.terminal = True  # Detiene integraci√≥n al alcanzar epsilon = 1
end_inflation.direction = 1    # Solo detecta de <1 a >1
```

### 2.2.2. Condiciones Iniciales
---
- Valor inicial del Campo:$$ \phi_0=15\,M_{Pl}, \quad \dot\phi_0 = \frac{-dV(\phi_0)}{3\,M^2_{Pl}}$$
- Aproximaci√≥n inicial de H utilizando la aproximaci√≥n Slow-Roll: $$ H_0 = \sqrt{\frac{V(\phi_0}{3\,M^2_{Pl}}}   $$-

Ecuaciones para $ \phi_0 $ seg√∫n el tipo de potencial

**1. Potencial Cuadr√°tico (quad)**
Para el potencial cuadr√°tico, el valor de $ \phi_0 $ se calcula a partir de una aproximaci√≥n anal√≠tica basada en el n√∫mero de e-folds $ N_{\text{star}} $ y el valor de $ \phi_{\text{end}} $ (fin de la inflaci√≥n).

- **Valor de $ \phi_{\text{end}} $:**
$$\phi_{\text{end}} = \sqrt{2} M_{\text{Pl}}$$

- **Argumento:**
$$\text{argument} = 4 N_{\text{star}} + \left( \frac{\phi_{\text{end}}}{M_{\text{Pl}}} \right)^2$$

- **Valor inicial $ \phi_0 $:**
$$\phi_0 = \sqrt{\text{argument}} \cdot M_{\text{Pl}} = \sqrt{4 N_{\text{star}} + \frac{\phi_{\text{end}}^2}{M_{\text{Pl}}^2}} \cdot M_{\text{Pl}}$$


**2. Potencial Cu√°rtico (quartic)**
Para el potencial cu√°rtico, el c√°lculo de $ \phi_0 $ sigue una forma similar, ajustada para el potencial de orden 4.

- **Valor de $ \phi_{\text{end}} $:**
$$\phi_{\text{end}} = 2 \sqrt{2} M_{\text{Pl}}$$

- **Argumento:**
$$\text{argument} = 8 N_{\text{star}} + \left( \frac{\phi_{\text{end}}}{M_{\text{Pl}}} \right)^2$$

- **Valor inicial $ \phi_0 $:**
$$\phi_0 = \sqrt{\text{argument}} \cdot M_{\text{Pl}} = \sqrt{8 N_{\text{star}} + \frac{\phi_{\text{end}}^2}{M_{\text{Pl}}^2}} \cdot M_{\text{Pl}}$$


**3. Potencial Hilltop (hilltop)**
Para el potencial hilltop, se utiliza una aproximaci√≥n anal√≠tica cerca del m√°ximo del potencial, basada en el par√°metro $ \mu $.

- **Valor de $ \phi_{\text{end}} $ (estimado con $ \epsilon_V \approx 1 $):**
$$\phi_{\text{end}} = \left( \frac{\mu^8}{32 M_{\text{Pl}}^2} \right)^{1/6}$$

- **Denominador:**
$$\text{denom} = 8 N_{\text{star}} M_{\text{Pl}}^2 + \frac{\mu^4}{\phi_{\text{end}}^2}$$

- **Valor inicial $ \phi_0 $:**
$$\phi_0 = \frac{\mu^2}{\sqrt{\text{denom}}} = \frac{\mu^2}{\sqrt{8 N_{\text{star}} M_{\text{Pl}}^2 + \frac{\mu^4}{\phi_{\text{end}}^2}}}$$


**4. Potencial Alpha (alpha)**
Para el potencial alpha, $ \phi_0 $ se calcula usando una aproximaci√≥n logar√≠tmica que depende del par√°metro $ \alpha $.

- **Valor de $ \phi_{\text{end}} $:**
$$\phi_{\text{end}} = \frac{4}{3 \alpha}$$

- **Par√°metro $ s $:**
$$s = \sqrt{\frac{2}{3 \max(\alpha, 10^{-12})}}$$

- **Valor inicial $ \phi_0 $:**
$$\phi_0 = \frac{1}{s} \ln \left( \frac{4 N_{\text{star}}}{3 \max(\alpha, 10^{-12})} \right)$$


**Notas:**

- $ N_{\text{star}} $: N√∫mero de e-folds, que representa la cantidad de expansi√≥n inflacionaria.
- $ M_{\text{Pl}} $: Masa de Planck, una constante f√≠sica fundamental.
- $ \mu $: Par√°metro del potencial hilltop, relacionado con la escala del potencial.
- $ \alpha $: Par√°metro del potencial alpha, que controla la forma del potencial.
- El t√©rmino $ \max(\alpha, 10^{-12}) $ en el potencial alpha evita divisiones por cero para valores peque√±os de $ \alpha $.
- Las ecuaciones usan PyTorch (por ejemplo, torch.sqrt, torch.as_tensor, torch.log) para c√°lculos num√©ricos con soporte para tensores.

```python
# Condiciones iniciales
def get_Initial_Conditions(param, phi0=None, Nstar = 60, kappa_time = 100):
    m = param['m']
    MPl = param['M_Pl']
    
    if phi0 is None:
        if param['potential']=='quad':
            phi_end = torch.sqrt(torch.as_tensor(2.0))* MPl
            argument = 4.0*Nstar + (phi_end/MPl)**2
            phi0 = torch.sqrt(torch.as_tensor(argument)) * MPl
        
        elif param['potential'] == 'quartic':   # œÜ0 ‚âà sqrt(8N + œÜ_end^2/M_Pl^2) M_Pl
            phi_end = 2.0*torch.sqrt(torch.as_tensor(2.0))*MPl
            argument = 8.0*Nstar + (phi_end/MPl)**2
            phi0 = torch.sqrt(torch.as_tensor(argument)) * MPl
                    
        elif param['potential'] == 'hilltop' :
            # Aproximaci√≥n anal√≠tica cerca de la cima (ver ¬ß2c)
            # N ‚âà (mu^4/(8 M_Pl^2)) (1/œÜ0^2 - 1/œÜ_end^2)  ‚áí  œÜ0 ‚âà mu^2 / sqrt(8 N M_Pl^2 + mu^4/œÜ_end^2)
            # Estimamos œÜ_end con Œµ_V‚âà1: œÜ_end ‚âà (mu**(8)/(32 M_Pl**2))**(1/6)
            phi_end = (param['mu']**8/(32.0*MPl**2))**(1.0/6.0)
            denom = 8.0*Nstar*(MPl**2) + (param['mu']**4)/(phi_end**2)
            phi0 =  (param['mu']**2)/torch.sqrt(torch.as_tensor(denom))
    
        elif param['potential'] == 'alpha':
            # œÜ0 ‚âà (1/s) ln(4N/(3Œ±)), s = sqrt(2/(3Œ±))
            phi_end = 4/(3*param['alpha'])
            argument=2.0/(3.0*max(param['alpha'],1e-12))
            s = torch.sqrt(torch.as_tensor(argument))
            phi0 =  (1.0/s)*torch.log(4.0*Nstar/(3.0*max(param['alpha'],1e-12)))
        elif param['potential'] == 'VNet':
            phi_end = 1.64
            phi0 = 5.0
    else:
        phi_end = torch.sqrt(torch.as_tensor(2.0))* MPl
        
    H0 = friedmann_Equation(phi0, 0.0, param)   # Aprox. inicial de H (slow-roll)
    phi_dot0 = -(param['c']**2)*dV_dphi(phi0, param) / (3 * H0) # Aprox. slow-roll para phi_dot    
    y0 = [phi0, phi_dot0]
    V0 = V(phi0,param)
    # Tiempo de integraci√≥n (en unidades M_Pl^-1)
    factor = param['hbar']/(param['m']*param['c']**2)
    t_span = (0, kappa_time*factor)
    t_eval = torch.linspace(0, kappa_time*factor, 10000)

    condition = {
        'phi0': phi0,
        'phi_dot0': phi_dot0,
        'phi_end': phi_end,
        'H0': H0,
        'y0': y0,
        'V0': V0,
        't_span': t_span,
        't_eval': t_eval
    }
    return condition

param['potential'] = 'hilltop'
condition = get_Initial_Conditions(param)
print('Condiciones Iniciales\n-----------------------')
print(f'phi0:\t  {condition["phi0"]:.3e}')
print(f'phi_dot0:  {condition["phi_dot0"]:.3e}')
print(f'H0:\t  {condition["H0"]:.3e}')
print(f'V0:\t  {condition["V0"]:.3e}')
print('Potencial:', param['potential'])
```

### 2.2.3. [M√âTODO 1. INTEGRACI√ìN EN TIEMPO] Soluci√≥n de las Ecuaciones Diferenciales. 
---
Pasos del M√©todo de Integraci√≥n

**1. Definici√≥n del Problema:**

- **Ecuaciones Diferenciales:** La funci√≥n equations proporciona el sistema de EDOs:
$$\frac{d}{dt} \begin{bmatrix} \phi \\ \dot{\phi} \end{bmatrix} = \begin{bmatrix} \dot{\phi} \\ -3 H \dot{\phi} - c^2 \frac{dV}{d\phi} \end{bmatrix}$$
donde $ \phi $ es el campo escalar, $ \dot{\phi} $ es su derivada temporal, $ H $ es el par√°metro de Hubble (calculado por friedmann_Equation), y $ \frac{dV}{d\phi} $ es la derivada del potencial.
- **Condiciones Iniciales:** El diccionario condition contiene:

- t_span: Intervalo de tiempo para la integraci√≥n, por ejemplo, $ [t_{\text{inicial}}, t_{\text{final}}] $.
- y0: Vector de condiciones iniciales $ [\phi_0, \dot{\phi}_0] $, donde $ \phi_0 $ es el valor inicial del campo y $ \dot{\phi}_0 $ su derivada inicial.
- t_eval: Puntos de tiempo opcionales donde se eval√∫a la soluci√≥n (si se especifica).
- phi0 y phi_end: Valores inicial y final (te√≥rico) del campo escalar, usados para diagn√≥stico.


- **Evento de Terminaci√≥n:** La funci√≥n events (por ejemplo, end_inflation) define la condici√≥n para detener la integraci√≥n, como cuando $ \epsilon = \min(\epsilon_V, \epsilon_H) = 1 $, indicando el fin de la inflaci√≥n.

**2. M√©todo de Integraci√≥n:**

La funci√≥n solve_ivp resuelve las EDOs utilizando el m√©todo especificado en el par√°metro method (por defecto, RK45). Este es un m√©todo de Runge-Kutta expl√≠cito de orden 4(5), que combina un m√©todo de cuarto orden con un control de error de quinto orden para ajustar el tama√±o del paso temporal adaptativamente.

**Par√°metros de solve_ivp:**

- fun: La funci√≥n equations, que define el sistema de EDOs.
- t_span: Intervalo de tiempo $ [t_{\text{inicial}}, t_{\text{final}}] $.
- y0: Condiciones iniciales $ [\phi_0, \dot{\phi}_0] $.
- args: Par√°metros adicionales pasados a equations, en este caso, el diccionario param (contiene constantes como $ \hbar $, $ c $, $ M_{\text{Pl}} $).
- t_eval: Puntos de tiempo donde se desea la soluci√≥n (opcional).
- events: Funci√≥n que define condiciones para detener la integraci√≥n (por ejemplo, end_inflation).


El m√©todo RK45 es robusto y eficiente para problemas no r√≠gidos, como los encontrados en modelos inflacionarios, ya que ajusta el paso temporal para mantener la precisi√≥n num√©rica.


**3. Resultados de la Integraci√≥n:**

La soluci√≥n sol contiene:

- sol.t: Tiempos en los que se evalu√≥ la soluci√≥n.
- sol.y[0]: Valores de $ \phi $ (phi).
- sol.y[1]: Valores de $ \dot{\phi} $ (phi_dot).
- sol.t_events: Tiempos en los que se activaron los eventos (por ejemplo, el fin de la inflaci√≥n).

A partir de estos resultados, se calculan:

- **Par√°metro de Hubble $ H $:**
$$H = \sqrt{\frac{\frac{1}{2} \dot{\phi}^2 + V(\phi)}{3}}$$
Nota: Esta expresi√≥n simplifica la ecuaci√≥n de Friedmann asumiendo ciertas unidades o normalizaciones (comparada con la versi√≥n anterior en friedmann_Equation).
- **Par√°metros de slow-roll:**

- $ \epsilon_V $: Calculado por epsilon_V_Equation.
- $ \epsilon_H $: Calculado por epsilon_H_Equation.
- $ \eta_V $: Calculado por eta_V_Equation.

-** N√∫mero de e-folds $ N $:**
- $$N = \int H \, dt$$
Calculado num√©ricamente usando el m√©todo del trapecio (cumulative_trapezoid) sobre $ H $ respecto al tiempo $ t $, con $ N(0) = 0 $.

```python
def integrate_inflation_by_time(equations, condition, param, events, method='RK45', debug = True):
    # Resolver las ecuaciones
    sol = solve_ivp(equations, condition['t_span'], condition['y0'], 
                    args=(param,), method=method, 
                    t_eval=condition['t_eval'], events=events,
                   )
    
    # Extraer resultados
    t = sol.t
    phi = sol.y[0]
    phi_dot = sol.y[1]
    
    H = friedmann_Equation(phi, phi_dot, param)
    #epsilon_V = 0.5 * phi_dot**2 / H**2
    epsilon_V = epsilon_V_Equation( phi, param)
    eta_V = eta_V_Equation( phi, param)
    epsilon_H = epsilon_H_Equation( H, phi_dot, param)
    
    # Calcular n√∫mero de e-folds
    N_fold = cumulative_trapezoid(H, t, initial=0)  # N_fold = int H dt
    
    # Imprimir resultados clave
    t_end = sol.t_events[0][0] if sol.t_events[0].size > 0 else t[-1]
    phi_end = sol.y[0][-1]
    N_total = N_fold[-1]
    if debug:
        print(f"Tiempo al fin de inflaci√≥n:        \t {t_end:.2f}")
        print(f"Valor del campo inicial:           \t {condition['phi0']:.3f}")
        print(f"Valor del campo al final (simul.): \t {phi_end:.3f}")
        print(f"Valor del campo al final (teorico):\t {condition['phi_end']:.3f}")
        print(f"N√∫mero total de e-folds:           \t {N_total:.2f}")

    results = {
        't' : t,                # Variable Temporal
        'phi' : phi,            # Campo Phi
        'phi_dot': phi_dot,     # Derivada del campo
        'H': H,                 # Par√°metro de Hubble
        'epsilon_V': epsilon_V,     # par√°metro epsilon V
        'epsilon_H': epsilon_H,     # par√°metro epsilon H
        'eta_V': eta_V,     # par√°metro eta V
        'N': N_fold,            # e-Folds          
        't_end': t_end,         # Tiempo Final
        'phi_end': phi_end,     # Valor final del campo      
        'N_total': N_total,     # n√∫mero total de e-Folds          
    }

    return results
    
param['potential'] = 'quad'
condition  =get_Initial_Conditions(param) 
results = integrate_inflation_by_time(equations, condition=condition, param=param, events=end_inflation, method='RK45')
```

### 2.3.1. Definici√≥n de Funciones H2, Epsilon_V, Epsilon_H y Eta_V en funcion del par√°metro N (e-Folds)
---
Estas funciones utilizan la variable $   u = \frac{d\phi}{dN}   $, donde $   \phi   $ es el campo escalar y $   N   $ es el n√∫mero de e-folds. 

**1. Cuadrado del Par√°metro de Hubble ($ H^2 $) (H2_from_phi_u)**
Esta funci√≥n calcula el cuadrado del par√°metro de Hubble $ H^2 $ en funci√≥n del campo escalar $ \phi $ y la derivada del campo respecto al n√∫mero de e-folds $ u = \frac{d\phi}{dN} $.

$$H^2 = \frac{V(\phi)}{3 M_{\text{Pl}}^2 - \frac{1}{2} u^2}$$

- **Descripci√≥n:**
Esta ecuaci√≥n deriva de la ecuaci√≥n de Friedmann en el contexto de la inflaci√≥n, expresada en t√©rminos de $ u = \frac{d\phi}{dN} $. Representa la tasa de expansi√≥n del universo ($ H^2 $) en funci√≥n del potencial $ V(\phi) $ y la derivada del campo respecto al n√∫mero de e-folds. El t√©rmino $ \frac{1}{2} u^2 $ refleja la contribuci√≥n de la energ√≠a cin√©tica del campo escalar, ajustada por la relaci√≥n entre el tiempo c√≥smico y el n√∫mero de e-folds ($ \frac{dN}{dt} = H $).

**2. Par√°metro de Slow-Roll $ \epsilon_V $ (epsV)**
Esta funci√≥n calcula el par√°metro de slow-roll basado en el potencial, $ \epsilon_V $, que mide la pendiente relativa del potencial.

$$\epsilon_V = \frac{1}{2} M_{\text{Pl}}^2 \left( \frac{V'}{V} \right)^2$$

- **Descripci√≥n:**
El par√°metro $ \epsilon_V $ mide la inclinaci√≥n del potencial $ V(\phi) $ normalizada por el propio potencial y la masa de Planck. Es un indicador clave de las condiciones de slow-roll: valores peque√±os ($ \epsilon_V \ll 1 $) implican que la inflaci√≥n puede sostenerse. La ecuaci√≥n es id√©ntica a la definida en epsilon_V_Equation de las funciones previas, pero adaptada para usar tensores de PyTorch.

**3. Par√°metro de Slow-Roll $ \eta_V $ (etaV)**
Esta funci√≥n calcula el par√°metro de slow-roll $ \eta_V $, relacionado con la curvatura del potencial.

$$\eta_V = M_{\text{Pl}}^2 \frac{V''}{V}$$

- **Descripci√≥n:**
El par√°metro $ \eta_V $ mide la curvatura del potencial (segunda derivada) relativa al potencial mismo, normalizada por $ M_{\text{Pl}}^2 $. Valores peque√±os de $ \eta_V $ ($ |\eta_V| \ll 1 $) son necesarios para mantener la inflaci√≥n en el r√©gimen de slow-roll. Esta ecuaci√≥n es consistente con la definida en eta_V_Equation de las funciones previas.

**4. Par√°metro de Slow-Roll $ \epsilon_H $ (epsH_from_u)**
Esta funci√≥n calcula el par√°metro de slow-roll $ \epsilon_H $ en funci√≥n de $ u = \frac{d\phi}{dN} $.

$$\epsilon_H = \frac{1}{2} \frac{u^2}{M_{\text{Pl}}^2}$$

- **Descripci√≥n:**
El par√°metro $ \epsilon_H $ mide la contribuci√≥n de la energ√≠a cin√©tica del campo escalar (representada por $ u^2 $) relativa a la escala de Planck. Dado que $ u = \frac{d\phi}{dN} = \frac{\dot{\phi}}{H} $ (porque $ \frac{dN}{dt} = H $), esta ecuaci√≥n es equivalente a:
$$\epsilon_H = \frac{1}{2} \frac{\dot{\phi}^2}{H^2 M_{\text{Pl}}^2}$$
Esto coincide con la definici√≥n de $ \epsilon_H $ en epsilon_H_Equation, pero expresada en t√©rminos de $ u $ en lugar de $ \dot{\phi} $ y $ H $. Valores peque√±os de $ \epsilon_H $ indican una expansi√≥n inflacionaria sostenida.

**Notas Generales**

- Uso de $ u = \frac{d\phi}{dN} $: La variable $ u $ representa la derivada del campo escalar respecto al n√∫mero de e-folds, lo que es √∫til en modelos inflacionarios porque $ N $ mide directamente la cantidad de expansi√≥n. La relaci√≥n entre $ u $ y la derivada temporal es $ u = \frac{\dot{\phi}}{H} $.
- **Estabilidad Num√©rica**: Las funciones usan torch.clamp para limitar denominadores cercanos a cero, evitando errores en c√°lculos num√©ricos (por ejemplo, $ \text{denom} \geq 10^{-30} $, $ V \geq 10^{-300} $).
- **Dependencia del Potencial:** Las funciones dependen de $ V(\phi) $, $ \frac{dV}{d\phi} $, y $ \frac{d^2V}{d\phi^2} $, calculadas externamente por $ V $, $ dV_dphi $, y $ d2V_dphi $. La forma espec√≠fica del potencial (cuadr√°tico, cu√°rtico, hilltop, alpha, etc.) determina los resultados.
- **Uso de PyTorch:** Las ecuaciones est√°n implementadas con tensores de PyTorch (torch.as_tensor, torch.clamp), lo que permite c√°lculos eficientes y compatibilidad con hardware como GPUs.
- **Constantes F√≠sicas:** El par√°metro param incluye $ M_{\text{Pl}} $ (masa de Planck), y otras constantes como $ \hbar $ o $ c $ pueden estar impl√≠citas en el potencial o en normalizaciones.

**Resumen**
Las ecuaciones proporcionadas calculan magnitudes clave para la din√°mica inflacionaria:

- $ H^2 $: Cuadrado del par√°metro de Hubble, que describe la tasa de expansi√≥n del universo.
- $ \epsilon_V $: Par√°metro de slow-roll basado en la pendiente del potencial.
- $ \eta_V $: Par√°metro de slow-roll basado en la curvatura del potencial.
- $ \epsilon_H $: Par√°metro de slow-roll basado en la energ√≠a cin√©tica del campo escalar.


```python
# ============================================================
# Magnitudes auxiliares: H(œÜ,u), Œµ_V, Œ∑_V, Œµ_H
# u = dœÜ/dN
# ============================================================

def H2_from_phi_u(phi, u, param):
    Mpl2  = torch.as_tensor(param['M_Pl'])**2
    V_    = V(phi, param)
    # H^2 = V / (3 Mpl^2 - 0.5 u^2)
    denom = 3.0 * Mpl2 - 0.5 * u**2
    # Evitar denominadores muy peque√±os
    denom = torch.clamp(denom, min=1e-30)
    return V_ / denom

def epsV(phi, param):
    Mpl2 = torch.as_tensor(param['M_Pl'])**2
    V_   = V(phi, param)
    Vp   = dV_dphi(phi, param)
    return 0.5 * Mpl2 * (Vp / torch.clamp(V_, min=1e-300))**2

def etaV(phi, param):
    Mpl2 = torch.as_tensor(param['M_Pl'])**2
    V_   = V(phi, param)
    Vpp   = d2V_dphi(phi, param)
    return Mpl2 * Vpp / torch.clamp(V_, min=1e-300)

def epsH_from_u(u, param):
    Mpl2 = torch.as_tensor(param['M_Pl'])**2
    return 0.5 * (u**2) / torch.clamp(Mpl2, min=1e-300)
```

#### 2.3.2. Condiciones Iniciales
---
Las funciones slowroll_u0 y slowroll_H0 calculan las condiciones iniciales aproximadas para la variable $ u_0 = \frac{d\phi}{dN} $ y el par√°metro de Hubble $ H_0 $ en el r√©gimen de slow-roll (rodadura lenta) para un modelo inflacionario, dado un valor inicial del campo escalar $ \phi_0 $. Estas funciones son esenciales para inicializar la integraci√≥n num√©rica de las ecuaciones diferenciales que describen la din√°mica inflacionaria, asumiendo que las condiciones de slow-roll son v√°lidas al inicio de la inflaci√≥n

**1. Condici√≥n Inicial de Slow-Roll para $ u_0 $ (slowroll_u0)**

Condici√≥n inicial $ u_0 $:
$$u_0 = \frac{d\phi}{dN} \approx - M_{\text{Pl}}^2 \frac{V'}{V}$$

- **Prop√≥sito:**

- La variable $ u = \frac{d\phi}{dN} $ representa la derivada del campo escalar $ \phi $ respecto al n√∫mero de e-folds $ N $, donde $ N $ mide la cantidad de expansi√≥n inflacionaria ($ \frac{dN}{dt} = H $).
- En el r√©gimen de slow-roll, se asume que la aceleraci√≥n del campo escalar ($ \ddot{\phi} $) es despreciable y que la fricci√≥n c√≥smica domina ($ 3H\dot{\phi} \approx -V' $). Esto lleva a la aproximaci√≥n:
$$\dot{\phi} \approx -\frac{V'}{3H}$$
Dado que $ u = \frac{d\phi}{dN} = \frac{\dot{\phi}}{H} $, se obtiene:
$$u \approx -\frac{V'}{3H^2}$$
Usando la ecuaci√≥n de Friedmann en slow-roll ($ H^2 \approx \frac{V}{3 M_{\text{Pl}}^2} $), se deriva:
$$u \approx - \frac{V'}{\frac{V}{M_{\text{Pl}}^2}} = - M_{\text{Pl}}^2 \frac{V'}{V}$$

- La funci√≥n slowroll_u0 calcula esta cantidad en $ \phi = \phi_0 $, proporcionando una condici√≥n inicial aproximada para $ u_0 $, que es la derivada inicial del campo respecto al n√∫mero de e-folds.

- Esta ecuaci√≥n se utiliza para inicializar la integraci√≥n num√©rica de las ecuaciones diferenciales en t√©rminos de $ N $ (en lugar de tiempo $ t $), ya que $ u = \frac{d\phi}{dN} $ es la variable relevante en formulaciones basadas en e-folds.
- La aproximaci√≥n es v√°lida cuando los par√°metros de slow-roll ($ \epsilon_V $, $ \eta_V $) son peque√±os, lo que garantiza una inflaci√≥n sostenida.


**2. Condici√≥n Inicial de Slow-Roll para $ H_0 $ (slowroll_H0)**

Condici√≥n inicial $ H_0 $:
$$H_0 = \sqrt{\frac{V}{3 M_{\text{Pl}}^2}}$$

- $ H_0 $ es la tasa de expansi√≥n inicial del universo en el modelo inflacionario, calculada bajo la aproximaci√≥n de slow-roll.

- Proporciona el valor inicial de $ H $, que determina la escala de la expansi√≥n c√≥smica y se usa para calcular otras cantidades, como el n√∫mero de e-folds $ N = \int H \, dt $.






```python
# ============================================================
# Inicializaci√≥n slow-roll (u0 y H0)
# ============================================================

def slowroll_u0(phi0, param):
    # u_SR = dœÜ/dN ‚âà - Mpl^2 V'/V
    Mpl2 = torch.as_tensor(param['M_Pl'])**2
    V_  = torch.as_tensor(V(phi0, param), dtype=torch.float64)
    Vp  = torch.as_tensor(dV_dphi(phi0, param), dtype=torch.float64)
    #print(phi0, V_, Vp, - Mpl2 * Vp / torch.clamp(V_, min=1e-300))
    return - Mpl2 * Vp / torch.clamp(V_, min=1e-300)

def slowroll_H0(phi0, param):
    Mpl2 = torch.as_tensor(param['M_Pl'])**2
    V_   = V(phi0, param)
    #print(phi0, V_, torch.sqrt(torch.clamp(V_ / (3.0 * Mpl2), min=1e-300)))
    return torch.sqrt(torch.clamp(V_ / (3.0 * Mpl2), min=1e-300))

```

### 2.3.3. Integraci√≥n utilizando Rung-Kutta de Orden 4 sobre el n√∫mero de e-Folders, N
---
 **Ecuaciones Diferenciales (rhs)**
La funci√≥n rhs define el sistema de ecuaciones diferenciales en t√©rminos del n√∫mero de e-folds $ N $, donde el estado del sistema es $ y = [\phi, u] $, con $ u = \frac{d\phi}{dN} $.

- **Variables:**

- $ \phi $: Campo escalar.
- $ u = \frac{d\phi}{dN} $: Derivada del campo escalar respecto al n√∫mero de e-folds.
- $ H $: Par√°metro de Hubble.
- $ V' = \frac{dV}{d\phi} $: Derivada del potencial respecto a $ \phi $.
- $ M_{\text{Pl}} $: Masa de Planck (de param['M_Pl']).

- **Ecuaciones:**

- Par√°metro de Hubble al cuadrado:
$$H^2 = \frac{V(\phi)}{3 M_{\text{Pl}}^2 - \frac{1}{2} u^2}$$

- Par√°metro de slow-roll $ \epsilon_H $:
$$\epsilon_H = \frac{1}{2} \frac{u^2}{M_{\text{Pl}}^2}$$

- Sistema de ecuaciones diferenciales:
$$\frac{d}{dN} \begin{bmatrix} \phi \\ u \end{bmatrix} = \begin{bmatrix} u \\ -(3 - \epsilon_H) u - \frac{\frac{dV}{d\phi}}{H^2} \end{bmatrix}$$

- **Derivaci√≥n:**

- La primera ecuaci√≥n, $ \frac{d\phi}{dN} = u $, es simplemente la definici√≥n de $ u $.
- La segunda ecuaci√≥n, $ \frac{du}{dN} $, se deriva de la ecuaci√≥n de Klein-Gordon en t√©rminos de $ N $

```python
# ============================================================
# Integraci√≥n en N con RK4 (estable y simple)
# Estado: y = [œÜ, u] con u = dœÜ/dN
# ============================================================

def rhs(phi, u, param):
    # y' = [u, du/dN]
    H2   = H2_from_phi_u(phi, u, param)
    H    = torch.sqrt(torch.clamp(H2, min=1e-300))
    eH   = 0.5 * u**2 / (torch.as_tensor(param['M_Pl'])**2)  # Œµ_H = u^2/(2 Mpl^2)
    # du/dN = -(3 - Œµ_H) u - V'/H^2
    Vp   = dV_dphi(phi, param)
    du   = -(3.0 - eH) * u - Vp / torch.clamp(H2, min=1e-300)
    return u, du, H

def integrate_inflation_by_efolds(
        param,
        condition,
        dN=1e-3,
        N_max=80.0,
        stop_when='epsV',   # 'epsV' o 'epsH'
        record_time=True,
        si_time=False,       # si True, devuelve t en segundos (requiere H en s^-1)
        debug = True,
        full_time = False
    ):
    # Inicializaci√≥n
    phi0 = torch.as_tensor(condition['phi0'])
    u0   = slowroll_u0(phi0, param)         # mejor arranque SR
    H0   = slowroll_H0(phi0, param)

    steps = int(N_max / dN) + 1
    N  = torch.zeros(steps, dtype=torch.float64)
    phi = torch.zeros_like(N); phi[0] = phi0
    phi_dot  = torch.zeros_like(N); phi_dot[0]  = u0
    H = torch.zeros_like(N); H[0] = H0

    epsilon_V = torch.zeros_like(N); epsilon_V[0] = epsV(phi0, param)
    eta_V = torch.zeros_like(N); eta_V[0] = etaV(phi0, param)
    epsilon_H = torch.zeros_like(N); epsilon_H[0] = epsH_from_u(u0, param)

    # tiempo acumulado
    t = torch.zeros_like(N)
    if record_time:
        # dt = dN / H ; si si_time=True y H en s^-1, t en segundos.
        # En Naturales/Planck, t sale en 1/(unidad de H). Conversi√≥n fuera.
        t[0] = torch.tensor(0.0)

    # Integraci√≥n RK4 en N
    stop_idx = steps-1
    
    for i in range(steps-1):
        phi_i = phi[i]
        u_i   = phi_dot[i]

        # k1
        f1_u, f1_du, H1 = rhs(phi_i, u_i, param)
        # k2
        f2_u, f2_du, _  = rhs(phi_i + 0.5*dN*f1_u, u_i + 0.5*dN*f1_du, param)
        # k3
        f3_u, f3_du, _  = rhs(phi_i + 0.5*dN*f2_u, u_i + 0.5*dN*f2_du, param)
        # k4
        f4_u, f4_du, H4 = rhs(phi_i + dN*f3_u,     u_i + dN*f3_du,     param)

        phi[i+1] = phi_i + (dN/6.0) * (f1_u + 2*f2_u + 2*f3_u + f4_u)
        phi_dot[i+1]  = u_i   + (dN/6.0) * (f1_du + 2*f2_du + 2*f3_du + f4_du)
        N[i+1]  = N[i] + dN

        # H al nuevo paso (mejor usar el √∫ltimo evaluado)
        H_new   = rhs(phi[i+1], phi_dot[i+1], param)[2]
        H[i+1] = H_new

        # Slow-roll params
        epsilon_V[i+1] = epsV(phi[i+1], param)
        eta_V[i+1] = etaV(phi[i+1], param)
        epsilon_H[i+1] = epsH_from_u(phi_dot[i+1], param)

        # tiempo
        if record_time:
            # dt = dN / H. Si quieres segundos en SI, aseg√∫rate que H est√© en s^-1 (ver m√°s abajo).
            dti = dN / torch.clamp(H_new, min=1e-300)
            t[i+1] = t[i] + dti

        # criterio de parada
        crit = epsilon_V[i+1] if stop_when == 'epsV' else epsilon_H[i+1]
        if crit >= 1.0 and not full_time:
            stop_idx = i+1
            break
    
    results = {
        't' : t[:stop_idx+1]*5.3e-44,                # tiempo
        'phi' : phi[:stop_idx+1],            # Campo Phi
        'phi_dot': phi_dot[:stop_idx+1],     # Derivada del campo (respecto de N)
        'H': H[:stop_idx+1],                 # Par√°metro de Hubble
        'epsilon_V': epsilon_V[:stop_idx+1],     # par√°metro epsilon V
        'epsilon_H': epsilon_H[:stop_idx+1],     # par√°metro epsilon H
        'eta_V': eta_V[:stop_idx+1],     # par√°metro eta V
        'N': N[:stop_idx+1],            # e-Folds          
        't_end': t[stop_idx].item()*5.3e-44,         # Tiempo Final
        'phi_end': phi[stop_idx].item(),       # Valor final del campo      
        'N_total': N[stop_idx].item(),     # n√∫mero total de e-Folds          
    }
    if debug:
        print(f"Tiempo al fin de inflaci√≥n:        \t {results['t_end']:.2e} s.")
        print(f"Valor del campo inicial:           \t {condition['phi0']:.3f}")
        print(f"Valor del campo al final (simul.): \t {results['phi_end']:.3f}")
        print(f"Valor del campo al final (teorico):\t {condition['phi_end']:.3f}")
        print(f"N√∫mero total de e-folds:           \t {results['N_total']:.2f}")
        
    return results
```

```python
# ============================================================
# Diagrama de barras
# ============================================================

x = torch.arange(len(potentials))  # posiciones en eje x
width = 0.35                    # ancho de barra

fig, ax1 = plt.subplots(figsize=(10, 6))

# Barras para n_s y r (con distinto eje)
ax2 = ax1.twinx()

# Paleta de colores
colors_ns = ['#3B82F6', '#2563EB', '#1D4ED8', '#1E40AF']  # tonos azules
colors_r  = ['#F59E0B', '#D97706', '#B45309', '#92400E']  # tonos naranjas

# Barras de n_s
bars1 = ax1.bar(x - width/2, n_s_list, width, color=colors_ns, label=r'$n_s$')

# Barras de r
bars2 = ax2.bar(x + width/2, r_list, width, color=colors_r, label=r'$r$')

# ============================================================
# Decoraci√≥n y ejes
# ============================================================

ax1.set_xticks(x)
ax1.set_xticklabels(potentials, fontsize=12)
ax1.set_xlabel('Tipo de potencial inflacionario', fontsize=13)
ax1.set_ylabel(r'√çndice espectral $n_s$', color='tab:blue', fontsize=13)
ax2.set_ylabel(r'Raz√≥n tensor-escalar $r$', color='tab:orange', fontsize=13)

# L√≠neas de referencia (Planck 2018)
ax1.axhline(0.965, color='k', linestyle='--', lw=1.5, label=r'Planck 2018 ($n_s$)')
ax1.fill_between([-0.5, len(potentials)-0.5], 0.965-0.004, 0.965+0.004, 
                 color='gray', alpha=0.2, label=r'$\pm 1\sigma$ rango Planck')

# L√≠mite superior t√≠pico de r (BICEP/Keck)
ax2.axhline(0.036, color='red', linestyle=':', lw=1.2, label='L√≠mite BICEP/Keck (2023)')

# Rango visual
ax1.set_ylim(0.9, 1.05)
ax2.set_ylim(0, 0.30)

# A√±ade texto con valores sobre cada barra
for b in bars1:
    height = b.get_height()
    ax1.text(b.get_x() + b.get_width()/2, height + 0.0003,
             f"{height:.3f}", ha='center', va='bottom', fontsize=10, color='navy')

for b in bars2:
    height = b.get_height()
    ax2.text(b.get_x() + b.get_width()/2, height + 0.003,
             f"{height:.3f}", ha='center', va='bottom', fontsize=10, color='darkorange')

# T√≠tulo y leyenda
plt.title('Comparaci√≥n de observables inflacionarios: $n_s$ y $r$', fontsize=15, pad=15)

# Leyendas combinadas
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper right', fontsize=10)

plt.tight_layout()
plt.show()
```
